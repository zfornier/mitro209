```python
## avant expliquer le pb de la linéarité et la nécessité de passer par des listes doublement chaînées
class Node :
    
    def __init__(self, N):
        #the purpose is to create a list of double-linked lists sorted by degree
        #it will facilitate the removing step and so we need to have access to the adds of the previous and of the next element
        self.previous = None
        self.next_ = None
        self.neighbour = []
        self.number = N #the nodes are named with integers
        self.degree = 0 #total number of edges which have N for one of its extremities
        #pas sûre d'avoir besoin du degré 
        #useful to have the order in which we will suppress the nodes to obtain the densest subgraph
        
        #addresses will be represented by int as if it was kind of indexes in the double-linked list

        #useful for the removing steps
    def setPrevious(self, node):
        self.previous = node
    def getPrevious(self):
        return self.previous
    def setNext(self,node):
        self.next_ = node
    def getNext(self):
        return self.next_
    def setNeighbour(self,elt):
        self.neighbour.append(elt)
    def getNeighbours(self):
        return self.neighbour
    def getNumber(self):
        return self.number
    def getDegree(self):
        return self.degree
    
    def remove(self):
        Pv = self.previous 
        Nx = self.next_
        if Pv==None:
            if Nx!=None:
                Nx.previous = None                
        elif Nx == None:
            Pv.next_ = None
        else:
            Pv.next_ = Nx
            Nx.previous = Pv
        
    def isHead(self):
        if self.previous == None:
            return True
        return False
                
    def degreeUpdate(self):
        self.degree = self.degree - 1 
            #the only thing we will do in the greedy algorithm is removing one edge for each neighbour of the considered node 
            #so the degree of the remaining nodes in the graph is either still the same or decreased of one
    
    def degreeInitialisation(self):
        self.degree = self.degree + 1
            #useful for the construction of the initial graph 
    
```

```python

#the graph is represented by an adjency list 
#for each node, we associate the list of its neighbours : a list of objects with type Nodes
class Graph : 
    
    def __init__(self, name_txt_file):
        #we could use a list of list where each index reprents a node and where there is the list of neighbours at the right index
        #but but this is not practical for defining subgraphs because their vertices do not form a sequence of consecutive integers.
        #Therefore, we use dictionnaries
        
        

        
        #*************************************** Reading of the text file ***************************************
        
        file=open(name_txt_file,'r')
        line = (file.readline().strip("\n")).split() #all the files I have elected have the size of the graph on 
        number_of_nodes = int(line[2]) #the 1st line
        number_of_edges = int(line[1])
        line = file.readline()
        
        E = [] #list of tuples that represents all the edges of the graph
        for k in range(number_of_edges):
            line=line.strip("\n").split()
            E.append([int(line[0]),int(line[1])])
            line = file.readline()
        file.close()
        
        #***************** initialisation of the graph and of the list of neighbours for each node**************
        self.edges = E
        self.nodes = [None for k in range(number_of_nodes)]
        self.numberNodes = number_of_nodes
        self.numberEdges = len(E)                
        self.degrees = [None for k in range(len(E))]
        self.density = len(E)/number_of_nodes 
            
        vertices = [] #list of the features of the vertices we add gradually
            
        for e in E :
            for i in range(2):
                if e[i] not in vertices:
                    vertices.append(e[i])
                    new_node = Node(e[i])
                    if e[i-1] not in vertices:
                        vertices.append(e[i-1])
                        self.nodes[e[i-1]-1] = Node(e[i-1])
                    new_node.setNeighbour(self.nodes[e[i-1]-1])
                    new_node.degreeInitialisation()
                    self.nodes[new_node.number-1]= new_node
                else:
                    exit_node = self.nodes[e[i]-1]
                    if e[i-1] not in vertices:
                        vertices.append(e[i-1])
                        self.nodes[e[i-1]-1] = Node(e[i-1])
                    exit_node.setNeighbour(self.nodes[e[i-1]-1])
                    exit_node.degreeInitialisation()
            
        #***************** initialisation of the graph and of the list of neighbours for each node**************
        
        def changeHead(self, d, new_head):
            headList=self.degrees[d]
            #We check the case where there were no node of degree d
            if headList==None:
                self.degrees[d]=new_head
                new_head.setNext(None)
                new_head.setPrevious(None)
            else:
                headList.setPrevious(new_head)
                new_head.setNext(headList)
                new_head.setPrevious(None)
                self.degrees[d]=new_head
                
       
        for n in self.nodes:
            d = n.getDegree()
            #if there is not yet node for this degree we add this one which will be the first 
            #of the double-linked list for this degree : the list of all nodes which have this degree
            changeHead(self,d, n)
    
        self.minimumDegree=None
        
        
    
    def updateMinimumDegree(self):
        #Initialisation: min=None, we want the first element!=None of degrees
        
        if self.minimumDegree==None:
            ind=0
            found=False
            while(not(found) and ind<self.numberEdges):
                if self.degrees[ind]!=None:
                    self.minimumDegree=self.degrees[ind]
                    found=True
                else:
                    ind+=1
        #There are 2 options: either the minimumDegree is minimumDegree - 1
        #in the case where we removed v, deg(v)>1, and one of v's neighbour
        #was of the same degree, either the minimumDegree is higher
        else: 
            d=self.minimumDegree.getDegree()
            if (self.degrees[d-1]!=None):
                self.minimumDegree=self.degrees[d-1]
            else:
                ind=d
                found=False
                while(not(found) and ind < self.numberEdges):
                    if self.degrees[ind]!=None:
                        self.minimumDegree=self.degrees[ind]
                        found=True
                    else:
                        ind+=1
                
        
    def getDensity(self):
        if self.numberNodes==0:
            return 0
        return self.numberEdges/self.numberNodes
    
    #This method adds a new node at the head of the list of nodes of degree d
    def changeHead(self, d, new_head):
        headList=self.degrees[d]
        #We check the case where there were no node of degree d
        if headList==None:
            self.degrees[d]=new_head
            new_head.setNext(None)
            new_head.setPrevious(None)
        else:
            headList.setPrevious(new_head)
            new_head.setNext(headList)
            new_head.setPrevious(None)
            self.degrees[d]=new_head
            
            
    #This method will update the list of heads degrees each time we remove 
    #a node from the graph, or we update its degree
    def removeNode(self, node):
        if (node.isHead()):
            d=node.getDegree()
            next_el=node.getNext() 
            #In the case where next_el=None, it is indeed the new head anyway 
            self.degrees[d]=next_el
        node.remove()
                
    
    #This method returns the densest-subgraph via the greedy algorithm
    def greedy(self):
        
        #G represents the states of each node: True means he has been removed
        G=[False for i in range(self.numberNodes)] 
        
        #H keeps track of the nodes we need to remove for a densest subgraph
        H=G.copy()
        
        while (self.numberNodes>0): 
            
            #print("Il nous reste " + str(self.numberNodes) + " à regarder")
            #We need to keep track of the nodes we remove in each step
            removedNodes=[]
            
            #We update the minimum degree
            self.updateMinimumDegree()
    
            #We look at the head of the list of nodes of minimum degree
            v=self.minimumDegree
            #print("The node of minimum degree is: " + str(v.number) + " of degree " + str(v.degree))
                
            #We remove it
            self.removeNode(v)  
            self.numberNodes-=1
            removedNodes.append(v.getNumber())
            G[v.getNumber()-1]=True 
                
            neighbours = v.getNeighbours() 

            #We update the degree of each neighbour                   
            for neighbour in neighbours:
                
                #print("On regarde le premier voisin de " + str(v.number) + " qui est :")
                number=neighbour.getNumber()
                #print(str(number) + " of degree " + str(neighbour.getDegree()))
                    
                #We check that this neighbour hasn't been removed 
                if G[number-1]==False:
                    
                    #print("This neighbour still exists")
                    #We remove neighbour from his current degree list
                    self.removeNode(neighbour)
                    neighbour.degreeUpdate()
                    degree=neighbour.getDegree()
                    self.numberEdges-=1 #the edge (v,neighbour) is removed
                    
                    #print("Its degree is now : " + str(neighbour.degree))
                    #If v was the only neighbour left of neighbour,
                    #then he will also be removed
                    if degree==0:
                        
                        #print("He has been erased")
                        G[number-1]=True
                        self.numberNodes-=1
                        #print("The number of nodes is : " + str(self.numberNodes))
                        removedNodes.append(number)
                            
                    else: #We add him as head of the list of his new degree
                        
                        #print("He has been added to the list of inferior degree")
                        self.changeHead(degree, neighbour)
            
            #Finally, we compute the density of the current subgraph
            current_density=self.getDensity()
            #print("The current density is " + str(current_density))

            if current_density>self.density:
                
                #print("We have found a subgraph")
                #G is a densest subgraph and we copy it in H
                for number in removedNodes:
                    H[number-1]=True
            
                    
        return H
    
    
       
```
